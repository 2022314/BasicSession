```python
headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/97.0.4692.71 Safari/537.36'}
```

```python
# 크롤링 할 대상 url 지정
url = 'https://www.saramin.co.kr/zf_user/jobs/list/domestic'

# 크롤링 대상 url과 headers를 기반으로 정보 요청
res = requests.get(url, headers = headers)

# 정상적으로 정보를 받은 경우 지속, 그렇지 않은 경우 중지
res.raise_for_status()

# html문서를 beautifulsoup의 lxml 형태로 변환
soup = BeautifulSoup(res.text, 'lxml') 
```

```python
# html문서에서 str_tit 라는 이름의 class를 지닌 a 태그의 element들 모두 선택
a1 = soup.findAll('a', attrs = {'class' : 'str_tit'})

# 회사 이름이 담겨있는 해당 html문서의 내용
'''
<div class="col company_nm">
        <a class="str_tit" onclick="s_trackApply(this, 'area_recruit', 'general');" title="손끝" href="/zf_user/jobs/relay/view?view_type=list&amp;rec_idx=41913979" target="_blank"><span>손끝</span></a>         <div class="toolTipWrap wrap_interested_corp"><button type="button" csn="ZVMrcWVvaFk5bzFDa2hGSW9ocFo4dz09" title="관심기업 등록" del_fl="n" class="interested_corp" onclick="try{Saramin.btnJob('favor', this, '', 'list');}catch(e){}" onmouseover="Saramin.favorTooltip(this, 'on');" onmouseout="Saramin.favorTooltip(this, 'off');" first_nudge="off"><span>관심기업 등록</span></button><div class="toolTip" style="display: none;"><span class="tail tail_bottom_center"></span><div class="toolTipCont txtCenter">관심기업 등록</div></div></div>                <div class="icon">
                    </div>
    </div>
'''

a1
```

```python
print(a1[0]['title'])
print(a1[1]['title'])

print(a1[2]['title'])
print(a1[3]['title'])
```

```python
#각 구인회사 이름과 내용을 추출할 반복문 생성
com_tit = []
job_tit = []
i = 0

while i < 80 :
  com_tit.append(a1[i]['title'])  # 회사이름 추가
  job_tit.append(a1[i+1]['title']) # 구직 내용 추가
  i += 2

print(' # of company in the list : {}'.format(len(com_tit)))  
print()
print('-'*80)
print()
print(com_tit)
print(job_tit)
```

```python
import pandas as pd

df = pd.DataFrame(com_tit,columns=['company name'])
df2 = pd.DataFrame(job_tit,columns=['job title'])
df_con = pd.concat([df,df2],axis=1)
df_con.index = range(1,41)

df_con
```
