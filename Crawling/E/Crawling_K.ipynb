```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/97.0.4692.71 Safari/537.36'}
```

```python
url='https://www.saramin.co.kr/zf_user/jobs/list/domestic'
```

```python
res=requests.get(url,headers=headers)
res.raise_for_status()
soup=BeautifulSoup(res.text,'lxml')
```

```python
companies=soup.find_all('li',attrs={'class':'item'})

n=0

list=[]
for company in companies:
    company_name=company.find('span',attrs={'class':'corp'}).get_text()
    title=company.find('a',attrs={'class':'link'})['title']
    print(f'{n+1}\n회사명:{company_name}\n제목:{title}')
    print()
    n+=1
    list.append([company_name,title])
    if n==40:
        break
```

```python
df=pd.DataFrame(list)
df.index=[i for i in range(1,41)]
df.columns=['회사명','제목']
df
```
